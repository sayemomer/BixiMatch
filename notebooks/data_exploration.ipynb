### Step:1- load the 2014 bike use data with weather

```python
import pandas as pd
import numpy as np

def load_data():
    data = pd.read_csv('/Users/sayems_mac/BixiMatch/data/OD/OD_2014_weather.csv')
    return data


df = load_data()

df.head()
```

### Step:2- Number of trips by weather description

```python
# graph the number of trips by weather condition

import matplotlib.pyplot as plt

df['Description'].value_counts().plot(kind='bar')
plt.show()
```

### Step:3- Number of trips by temperature

```python
# graph the temperature to number of trips distribution

df['Temperature_C'].hist(bins=20)

plt.show()

```

### Step:4- Number of trips by wind speed


```python
# graph the number of trips by precipitation

df['Wind_speed'].hist(bins=20)

plt.show()
```

### Step:5- Number of trips by weekday

```python
# graph the number of trips by days of the week



df['weekday'] = pd.to_datetime(df['start_date']).dt.weekday

df['weekday'].value_counts().plot(kind='bar')

plt.show()
```

```python
# weekday vs weekend

df['weekend'] = df['weekday'].apply(lambda x: 1 if x >= 5 else 0)

df['weekend'].value_counts().plot(kind='bar')

plt.show()
```

### Step:6- Number of trips by station

```python
# graph the number of trips by station

df['start_station_code'].value_counts().plot(kind='hist')

plt.show()

```

### Step:7- Time series analysis

```python
# time series analysis

df['start_date'] = pd.to_datetime(df['start_date'])

df.set_index('start_date', inplace=True)

df['start_station_code'].resample('D').count().plot()

plt.show()
```

```python
# graph the number of trips by temperature

df['Temperature_C'].resample('D').mean().plot()

plt.show()
```
